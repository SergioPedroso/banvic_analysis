```python
qtd_por_dia = (
    df_obt_transacoes.groupby(["date_day","week_day"])
      .agg(qtd_transacoes=("transacao_id","count"),
           valor_total=("valor_transacao","sum"))
      .reset_index()
)

media_qtd = qtd_por_dia.groupby("week_day")["qtd_transacoes"].mean().reset_index()
media_valor = qtd_por_dia.groupby("week_day")["valor_total"].mean().reset_index()


ordem_dias = ["Segunda-feira","Terça-feira","Quarta-feira","Quinta-feira",
              "Sexta-feira","Sábado","Domingo"]

media_qtd["week_day"] = pd.Categorical(media_qtd["week_day"], categories=ordem_dias, ordered=True)
media_valor["week_day"] = pd.Categorical(media_valor["week_day"], categories=ordem_dias, ordered=True)

plt.figure(figsize=(8,5))
sns.barplot(data=media_qtd, x="week_day", y="qtd_transacoes", color="blue")
plt.title("Média diária de transações por dia da semana")
plt.xlabel("Dia da semana")
plt.ylabel("Média de transações")
plt.grid(False, axis='x')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=media_valor, x="week_day", y="valor_total", color="red")
plt.title("Média diária de volume movimentado por dia da semana")
plt.xlabel("Dia da semana")
plt.ylabel("Média de volume movimentado (R$)")
plt.xticks(rotation=45)
plt.grid(False, axis='x')
plt.show()
```

```python
from scipy.stats import ttest_ind

df_obt_transacoes["month_num"] = pd.to_datetime(df_obt_transacoes["date_day"]).dt.month
df_obt_transacoes["year_number"] = pd.to_datetime(df_obt_transacoes["date_day"]).dt.year

volume_mensal = df_obt_transacoes.groupby(["year_number","month_num","month_name_short"])["valor_transacao"].sum().reset_index()
volume_mensal["par"] = volume_mensal["month_num"] % 2 == 0
volume_mensal["grupo"] = volume_mensal["par"].map({True:"Par", False:"Ímpar"})

plt.figure(figsize=(6,5))
sns.boxplot(data=volume_mensal, x="grupo", y="valor_transacao", palette="pastel")
plt.title("Distribuição de volume mensal\nMeses pares vs ímpares")
plt.xlabel("")
plt.ylabel("Volume mensal movimentado (R$)")
plt.show()

media_por_mes = volume_mensal.groupby("month_num")["valor_transacao"].mean().reset_index()
media_por_mes["month_name_short"] = pd.to_datetime(media_por_mes["month_num"], format="%m").dt.strftime("%b")

plt.figure(figsize=(10,5))
sns.barplot(data=media_por_mes, x="month_name_short", y="valor_transacao", color="blue")
plt.title("Volume médio mensal por mês do ano")
plt.xlabel("Mês")
plt.ylabel("Volume médio movimentado (R$)")
plt.show()

pares = volume_mensal.loc[volume_mensal["par"],"valor_transacao"]
impares = volume_mensal.loc[~volume_mensal["par"],"valor_transacao"]

t_stat, p_value = ttest_ind(pares, impares, equal_var=False)

print(f"Teste t (Welch): t={t_stat:.3f}, p={p_value:.4f}")
if p_value < 0.05:
    print("Hipótese nula rejeitada. Há diferença estatística entre as médias.")
else:
    print("Hipótese nula aceita. Não há evidência de diferença estatística entre médias.")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

df_medianas = (
    df_clientes
    .merge(df_contas, left_on='cod_cliente', right_on='cod_cliente')
    .groupby('faixa_idade')[['saldo_total', 'saldo_disponivel']]
    .median()
    .reset_index()
)
df_medianas_melt = df_medianas.melt(id_vars='faixa_idade', value_vars=['saldo_total', 'saldo_disponivel'],
                                    var_name='Tipo de Saldo', value_name='Valor (R$)')

df_saldos_faixa = (
    df_clientes
    .merge(df_contas, left_on='cod_cliente', right_on='cod_cliente')
    .groupby('faixa_idade')[['saldo_total', 'saldo_disponivel']]
    .sum()
    .reset_index()
)
df_saldos_faixa_melt = df_saldos_faixa.melt(id_vars='faixa_idade', value_vars=['saldo_total', 'saldo_disponivel'],
                                            var_name='Tipo de Saldo', value_name='Valor (R$)')

sns.barplot(
    data=df_medianas_melt,
    x='faixa_idade', y='Valor (R$)', hue='Tipo de Saldo', palette='colorblind', ax=axes[0]
)
axes[0].set_title('Mediana do Saldo Total e Disponível por Faixa Etária')
axes[0].set_xlabel('Faixa Etária')
axes[0].set_ylabel('Mediana (R$)')
axes[0].grid(False)

sns.barplot(
    data=df_saldos_faixa_melt,
    x='faixa_idade', y='Valor (R$)', hue='Tipo de Saldo', palette='colorblind', ax=axes[1]
)
axes[1].set_title('Soma do Saldo Total e Disponível por Faixa Etária')
axes[1].set_xlabel('Faixa Etária')
axes[1].set_ylabel('Soma (R$)')
axes[1].grid(False)

plt.tight_layout()
plt.show()
```

### Análise feriados

```python
df_feriados = pd.read_csv('../data/feriados_pontes_brasil_2010_2025.csv')
```

```python
df_feriados.head(2)
```

```python
df_transacoes["data"] = pd.to_datetime(df_transacoes["data_transacao"], errors="coerce").dt.date
df_feriados["date"]   = pd.to_datetime(df_feriados["date"], errors="coerce").dt.date

feriados = set(df_feriados.loc[df_feriados["type"].astype(str).str.lower().str.contains("holiday"), "date"])
pontes   = set(df_feriados.loc[df_feriados["type"].astype(str).str.lower().str.contains("bridge"),  "date"])

def classifica(d):
    if d in feriados: return "Feriado"
    if d in pontes:   return "Ponte"
    return "Final de semana" if pd.Timestamp(d).weekday() >= 5 else "Dia útil"

df_transacoes = df_transacoes.dropna(subset=["data"]).copy()
df_transacoes["tipo_dia"] = df_transacoes["data"].apply(classifica)

daily = (df_transacoes
         .groupby(["data","tipo_dia"], as_index=False)
         .agg(volume=("data","count"), valor_total=("valor_transacao","sum")))

avg = (daily.groupby("tipo_dia", as_index=False)
             .agg(media_volume=("volume","mean"), media_valor=("valor_total","mean"))
             .sort_values("tipo_dia"))

print(avg)

x = np.arange(len(avg))

plt.figure(figsize=(7,4))
plt.bar(x, avg["media_volume"])
plt.xticks(x, avg["tipo_dia"])
plt.title("Média diária de volume por tipo de dia")
#plt.xlabel("Tipo de dia") 
plt.ylabel("Transações/dia")
#for i,v in enumerate(avg["media_volume"]): plt.text(i, v, f"{v:.1f}", ha="center", va="bottom")
plt.grid(False, axis='x')
plt.ylim(0, 1.1*avg["media_volume"].max())
plt.tight_layout()
plt.show()

plt.figure(figsize=(7,4))
plt.bar(x, avg["media_valor"])
plt.xticks(x, avg["tipo_dia"])
plt.title("Média diária de valor por tipo de dia")
#plt.xlabel("Tipo de dia") 
plt.ylabel("Valor/dia")
#for i,v in enumerate(avg["media_valor"]): plt.text(i, v, f"{v:,.0f}", ha="center", va="bottom")
plt.grid(False, axis='x')
plt.ylim(0, 1.1*avg["media_valor"].max())
plt.tight_layout()
plt.show()
```
